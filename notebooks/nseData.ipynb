```python
# download nse data for backtesting
from yfinance import download

# (Yahoo error = "1h data not available for startTime=1577817000 and endTime=1717180200. The requested range must be within the last 730 days.")'
tickers = ["RELIANCE.NS", "TCS.NS", "HDFCBANK.NS", "INFY.NS", "ICICIBANK.NS"]
data = download(tickers, period="730d", interval="1h")
data.to_csv("data/nse_1h.csv")

    
```

```python
# download all stocks available in nse
from nsepy import get_stock_codes
```

```python
# create a new column 'pct_change' in the dataframe that contains the percentage change of the 'Close' price for each stock
for ticker in tickers:
    data[('Pct_Change', ticker)] = data[('Close', ticker)].pct_change().fillna(0)
```
